### Import Necessary Classes

```python
import pandas as pd
import numpy as np
import os
import json
import datetime
```

### Read json data

```python
directory = r'C:\Users\govin\Documents\Anaconda\Data'
def read_file(filename):
    file=open(os.path.join(directory,filename))
    data=[json.loads(l) for l in file]
    dataframe=pd.json_normalize(data)
    return dataframe
```

```python
brand=read_file('brands.json')
```

```python
users=read_file('users.json')
receipts=read_file('receipts.json')
```
